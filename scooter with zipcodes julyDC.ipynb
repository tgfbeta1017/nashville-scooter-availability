```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
import itertools
```

```python
import pickle


with open('data/july.pkl', 'rb') as f:
    july_data = pickle.load(f)
```

```python
july_data.head(1)
```

```python
zipcode = gpd.read_file('./data/zipcode.geojson')
```

```python
nash= zipcode[(zipcode['po_name'])=='NASHVILLE']
```

```python
july_scooter_geo = gpd.GeoDataFrame(july_data,crs=nash.crs,geometry=july_data['start_geo'])
```

```python
july_scooter_geo.head()
```

```python
nash_july_scooter=gpd.sjoin(july_scooter_geo,nash,op='within')
```

```python
nash_july_scooter.head(3)
```

```python
july_scooter_zipcodes=nash_july_scooter[['sumdID', 'zip']].groupby('zip').agg('count').sort_values(by ='zip', ascending = False)
```
